```python
import pandas as pd

data = [
    ["Amul Milk 1L", "Dairy", 60, "Fresh toned milk"],
    ["Aavin Milk 1L", "Dairy", 58, "Standard milk"],
    ["Brown Bread", "Bakery", 40, "Whole wheat bread"],
    ["White Bread", "Bakery", 35, "Soft sandwich bread"],
    ["Eggs 12 pack", "Dairy", 90, "Farm fresh eggs"],
    ["Banana 1 dozen", "Fruits", 50, "Fresh bananas"],
    ["Apple 1kg", "Fruits", 120, "Shimla apples"],
    ["Oats 500g", "Breakfast", 150, "Healthy oats"],
    ["Cornflakes 500g", "Breakfast", 180, "Crunchy cereal"],
    ["Rice 5kg", "Staples", 350, "Ponni rice"],
    ["Wheat Flour 5kg", "Staples", 250, "Atta"],
    ["Sugar 1kg", "Staples", 45, "Refined sugar"],
    ["Salt 1kg", "Staples", 20, "Iodized salt"],
    ["Sunflower Oil 1L", "Cooking", 140, "Refined oil"],
    ["Toor Dal 1kg", "Staples", 160, "Protein rich dal"]
]

df = pd.DataFrame(data, columns=["product_name", "category", "price", "description"])
df.head()
```

```python
from sentence_transformers import SentenceTransformer
import chromadb

model = SentenceTransformer('all-MiniLM-L6-v2')

texts = df['product_name'] + " " + df['category'] + " " + df['description']
embeddings = model.encode(texts.tolist())

client = chromadb.Client()
collection = client.create_collection("groceries")

for i in range(len(df)):
    collection.add(
        documents=[texts.iloc[i]],
        embeddings=[embeddings[i]],
        metadatas=[df.iloc[i].to_dict()],
        ids=[str(i)]
    )

print("✅ Vector DB ready")
```

```python
cart = []

def search_groceries(query):
    results = collection.query(query_texts=[query], n_results=5)
    return results['metadatas'][0]

def add_to_cart(item):
    cart.append(item)
    return f"✅ Added {item['product_name']}"

def calculate_total():
    return sum(item['price'] for item in cart)

def show_cart():
    for item in cart:
        print(f"{item['product_name']} - ₹{item['price']}")
    print(f"\nTotal: ₹{calculate_total()}")
```

```python
def agent(user_input):

    user_input = user_input.lower()

    # Budget shopping
    if "under" in user_input:
        numbers = [int(s) for s in user_input.split() if s.isdigit()]
        budget = numbers[0] if numbers else 500

        items = search_groceries(user_input)

        selected = []
        total = 0

        for item in items:
            if total + item['price'] <= budget:
                selected.append(item)
                total += item['price']

        return selected, f"🧾 Total: ₹{total}"

    # Add item
    elif "add" in user_input:
        items = search_groceries(user_input)
        return add_to_cart(items[0])

    # Show total
    elif "total" in user_input:
        return f"🧾 Total: ₹{calculate_total()}"

    # Default search
    else:
        return search_groceries(user_input)
```

```python
from typing import TypedDict, List, Dict

class AgentState(TypedDict):
    input: str
    intent: str
    results: List[Dict]
    cart: List[Dict]
    total: float
    budget: float
    response: str
```

```python
state = {
    "input": "",
    "intent": "",
    "results": [],
    "cart": cart,   # reuse your existing cart
    "total": 0,
    "budget": 0,
    "response": ""
}
```

```python
def planner_node(state):

    user_input = state["input"].lower()

    if "under" in user_input:
        return {"intent": "budget"}
    elif "add" in user_input:
        return {"intent": "add"}
    elif "total" in user_input:
        return {"intent": "total"}
    else:
        return {"intent": "search"}
```

```python
def search_node(state):
    results = search_groceries(state["input"])
    return {"results": results, "response": results}
```

```python
def add_node(state):
    item = search_groceries(state["input"])[0]
    msg = add_to_cart(item)
    return {"response": msg}
```

```python
def total_node(state):
    total = calculate_total()
    return {"total": total, "response": f"🧾 Total: ₹{total}"}
```

```python
def budget_node(state):

    user_input = state["input"]

    numbers = [int(s) for s in user_input.split() if s.isdigit()]
    budget = numbers[0] if numbers else 500

    items = search_groceries(user_input)

    selected = []
    total = 0

    for item in items:
        if total + item['price'] <= budget:
            selected.append(item)
            total += item['price']

    return {
        "budget": budget,
        "results": selected,
        "total": total,
        "response": f"🛒 Basket under ₹{budget}: {selected} | Total ₹{total}"
    }
```

```python
from langgraph.graph import StateGraph

graph = StateGraph(AgentState)

# Add nodes
graph.add_node("planner", planner_node)
graph.add_node("search", search_node)
graph.add_node("add", add_node)
graph.add_node("total", total_node)
graph.add_node("budget", budget_node)

# Entry point
graph.set_entry_point("planner")
```

```python
def route(state):
    return state["intent"]

graph.add_conditional_edges(
    "planner",
    route,
    {
        "search": "search",
        "add": "add",
        "total": "total",
        "budget": "budget"
    }
)
```

```python
app = graph.compile()
```

```python
def run_agent(query):
    state["input"] = query
    result = app.invoke(state)
    return result["response"]
```

```python
import os
from groq import Groq

# os.environ["GROQ_API_KEY"] = "YOUR_API_KEY" # This line was removed/commented out

client = Groq()

def call_llm(prompt):
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant", # Suggested common Groq model
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )
    return response.choices[0].message.content
```

```python
from typing import TypedDict, List, Dict

class AgentState(TypedDict):
    input: str
    intent: str
    tool: str
    tool_input: str
    results: List[Dict]
    cart: List[Dict]
    total: float
    budget: float
    history: List[str]
    response: str
```

```python
def planner_node(state):

    prompt = f"""
    You are a grocery shopping assistant.

    User query: {state["input"]}

    Available tools:
    1. search_groceries
    2. add_to_cart
    3. calculate_total
    4. budget_optimizer

    Decide:
    - tool name
    - what input to pass

    Output JSON:
    {{
        "tool": "...",
        "tool_input": "..."
    }}
    """

    output = call_llm(prompt)

    try:
        import json
        parsed = json.loads(output)
    except:
        parsed = {"tool": "search_groceries", "tool_input": state["input"]}

    return {
        "tool": parsed["tool"],
        "tool_input": parsed["tool_input"]
    }
```

```python
def tool_node(state):

    tool = state["tool"]
    inp = state["tool_input"]

    if tool == "search_groceries":
        results = search_groceries(inp)
        return {"results": results}

    elif tool == "add_to_cart":
        item = search_groceries(inp)[0]
        msg = add_to_cart(item)
        return {"response": msg}

    elif tool == "calculate_total":
        total = calculate_total()
        return {"total": total}

    elif tool == "budget_optimizer":
        numbers = [int(s) for s in inp.split() if s.isdigit()]
        budget = numbers[0] if numbers else 500

        items = search_groceries(inp)

        selected = []
        total = 0

        for item in items:
            if total + item["price"] <= budget:
                selected.append(item)
                total += item["price"]

        return {"results": selected, "total": total}

    return {"response": "Tool not found"}
```

```python
def response_node(state):

    prompt = f"""
    You are a helpful grocery assistant.

    User query: {state["input"]}

    Results: {state.get("results", [])}
    Cart: {state.get("cart", [])}
    Total: {state.get("total", 0)}

    Generate a helpful, natural response.
    """

    response = call_llm(prompt)

    return {"response": response}
```

```python
from langgraph.graph import StateGraph

graph = StateGraph(AgentState)

graph.add_node("planner", planner_node)
graph.add_node("tool", tool_node)
graph.add_node("response", response_node)

graph.set_entry_point("planner")

graph.add_edge("planner", "tool")
graph.add_edge("tool", "response")
```

```python
app = graph.compile()
```

```python
state = {
    "input": "",
    "intent": "",
    "tool": "",
    "tool_input": "",
    "results": [],
    "cart": cart,
    "total": 0,
    "budget": 0,
    "history": [],
    "response": ""
}
```

```python
def run_agent(query):

    state["input"] = query
    state["history"].append(query)

    result = app.invoke(state)

    return result["response"]
```
